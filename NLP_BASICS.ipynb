NLP BASICS



```python
#text-> lowercase()
T="I LOVE LEARNing it nlp !! IT's very interesting?"
t1=T.lower()
t1
```

```python
#text-> remove punctation
import re
text_clean=re.sub(r'[^a-zA-Z\s]','',t1)
text_clean
```

```python
#text -> split by space
token=text_clean.split()
token
```

```python
#text -> stemming and lemmating to improve text be meaningful word
from nltk.stem import PorterStemmer
st=PorterStemmer()
st_word=[st.stem(word) for word in token]
st_word
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
l1=WordNetLemmatizer()
l1_words=[l1.lemmatize(word) for word in token]
l1_words
```

```python
final=" ".join(l1_words)
final
```

```python
#text->numbers bag of words  and tf-idf
from sklearn.feature_extraction.text import CountVectorizer
c=CountVectorizer()
x=c.fit_transform([final])
print(c.get_feature_names_out())
print(x.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
y=tf.fit_transform([final])
print(tf.get_feature_names_out())
print(y.toarray())
```
